```python
def mergeLists(keys, values):
    if len(keys) == len(values):
        new_dict = {}   # Create dictionary BEFORE the loop
        for i in range(len(values)):
            a = keys[i]
            b = values[i]
            new_dict[a] = b  # Add key/value pair
        return new_dict
    else:
        print("Lists do not match in length")
        return {}
```

```python
def compress_consecutive(lst):
	if len(lst) == 0:
		return []

	new_lst = [lst[0]]
	count = 0
	for i in range(len(lst)-1):
		if new_lst[count] == lst[i]:
			continue
		if lst[i] == lst[i+1]:
			new_lst.append(lst[i])      # Append is a fuction -> ()
			count += 1
		else:
			new_lst.append(lst[i])
	
	return new_lst

compress_consecutive([1, 1, 2, 2,2,2, 2, 3, 1, 1])
```

```python
def compress_consecutive(lst):
    if len(lst) == 0:
        return []

    new_lst = [lst[0]]  # start with first element

    for i in range(len(lst)-1):
        if lst[i] != lst[i+1]:      # only append when a NEW run starts
            new_lst.append(lst[i+1])

    return new_lst

compress_consecutive([1, 1, 2, 2, 2, 3, 1, 1])
```

```python
def alternate_sum(lst):
    s = 0
    for i in range(len(lst)):
        if i % 2 == 0:    #0,2,4.. even-numbers
            s+= lst[i]  #subtracting even numbers
        else:
            s-=lst[i]
    return s

alternate_sum([1,2,3,4])
```

```python
def count_local_maxima(lst):
	counter = 0
	for i in range(1, len(lst)-1):
		if lst[i] > lst[i-1] and lst[i]> lst[i+1]:
			counter += 1
			
	return counter

count_local_maxima([5, 4, 5, 2, 1])
```

```python
def build_frequency_dict(lst):
	new_dict = {}
	for i in lst:
		count = 0
		for j in lst:
			if i==j:
				count+=1
		new_dict[i] = count
	
	return new_dict

build_frequency_dict([1, 2, 2, 3, 1, 3, 3])
#{1: 2, 2: 2, 3: 3}
```

```python
#faster runtime-way
def build_frequency_dict(lst):
    new_dict = {}
    for i in lst:
        if i in new_dict:
            new_dict[i] += 1
        else:
            new_dict[i] = 1
    return new_dict

build_frequency_dict([1, 2, 2, 3, 1, 3, 3])
```
