Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026]) #faster than python lists
y1 = np.array([15, 25, 30, 20])
y2 = np.array([30, 15, 10, 25])

plt.plot(x, y1,marker=".", 
               markersize = 20, #makes dots on reported values
               markerfacecolor = "#48D54C", #changes the color
               color = "#199AC5", #color of the graph, either "blue" or hex-decimal
               linewidth = 3, #width of the graph
               linestyle = ":", #style of the graph
               markeredgecolor = "green" #edge color of the markers
)
plt.plot(x, y2,marker="*", 
               markersize = 20, #makes dots on reported values
               markerfacecolor = "#48D54C", #changes the color
               color = "#A429BA", #color of the graph, either "blue" or hex-decimal
               linewidth = 3, #width of the graph
               linestyle = "--", #style of the graph
               markeredgecolor = "blue" #edge color of the markers
)
```

```python
#use ditionaries, if you only want to change colors etc

line_styles = dict(marker=".", 
               markersize = 20, #makes dots on reported values
               markerfacecolor = "#48D54C", #changes the color
               linewidth = 3, #width of the graph
               linestyle = ":", #style of the graph
               markeredgecolor = "green" #edge color of the markers)
)

plt.plot(x, y1, color = "#199AC5", **line_styles) # ** to unpack the dictionary
plt.plot(x,y2, color = "#42D345", **line_styles)
```

Labels

```python
#customize label, titles, ticks

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20, 30])

plt.title("Class size", color="#8E7F08",
                        fontsize=20,
                        family="Arial",
                        fontweight="bold",
                         )
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xticks(x) #only ticks at the points of measure
plt.tick_params(axis="both",
                colors = "#3E2F13B7")

plt.xlabel("Year", fontsize = 12,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#0D4993B7")
plt.ylabel("Students", fontsize = 12,
                   family = "Arial",
                   fontweight = "bold",
                   color = "#430D93B7")
```

Pie charts

```python
#circular chart divided into slices to show a total. visualization of distribution among categories

categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])
colors = ["red", "blue", "green", "yellow"]

plt.pie(values, labels = categories,
        autopct="%1.1f%%", # displaying percentages and %% to add pct sign
        colors = colors,
        explode=[0,0,0,0.2], #takes the slige 0.2 out of the chart
        shadow=True, #shadow behind the chart
        startangle= 90 #rotating the chart
        )
plt.title("Einars Collage")


hi = 1 #just something at the bottom to avoid garbage
```

Histograms

```python
#visual representation of the distribution
#they group values into intervals (bins), and count how many fall into each range
#great for statistics and popability

test_scores = np.array([np.random.randint(1,100) for i in range(100)])
test_scores_advanced = np.random.normal(loc=80, #location of center: cluster around this score
                                        scale=10, #standard deviation, the spread, higher the number -> greater deviation
                                        size=100) #how many scores
test_scores_advanced = np.clip(test_scores_advanced, 0, 100 ) #setting the bounds: all scores outside range will be set to closest limit

plt.hist(test_scores_advanced,
         bins = 10, #how many bins (intervals)
         color = "Green",
         edgecolor = "black"
         )
plt.title("Exam scores")
plt.xlabel("Score")
plt.ylabel("Number of students")
```

Subplots

```python
#figure is the enire canvas
# Ax = A single plot (subplot)

x = np.array([i for i in range(20)])

figure, axes = plt.subplots(2, 2, figsize = (6,5)) #number of plots, with rows and columns (like a matrix)
                                                   #figure is the canvas

axes[0,0].plot(x,x*2, color = "Red")
axes[0,0].set_title("x*2")

axes[0,1].plot(x,x**2, color = "green")
axes[0,1].set_title("x**2")


axes[1,0].plot(x,x**(1/2), color = "Blue")
axes[1,0].set_title("root of x")

axes[1,1].plot(x,np.log(x), color = "Pink")
axes[1,1].set_title("ln(x)")


plt.tight_layout() #so that everything fits
plt.show()


```

Pandas with matplotlib

```python
import pandas as pd
df = pd.read_csv("pokemon.csv")

type_count = df["Type1"].value_counts().sort_values() #creates a series of type1 as index and values as column 

#bar chart with types and quantity
plt.barh(type_count.index, type_count.values, color = "pink", edgecolor = "black")
plt.title("Number of pokemon by primary type")
plt.xlabel("count")
plt.ylabel("Type")

plt.tight_layout()
plt.show()
```
